```python
print()
print('----------------------------------------------RAM POTHINENI--------------------------------------------------------')
print()

tweet_ram = ramTweets[0]
print(tweet_ram.text)

tweet_ram  = ramTweets[0]
print(tweet_ram.source)
print()
tokens_ram = tweet_ram.text.split()
print('Tokens after spliting: ', tokens_ram)
print('Length: ', len(tokens_ram))
print()

#make everything upper or lower case
lower_tokens_ram = [s.lower() for s in tokens_ram]
print('Lower case: ', lower_tokens_ram)
print()

#finding out unique tokens
unique_lower_ram = set(lower_tokens_ram)
print('Unique lower tokens:  ', unique_lower_ram)
print('Unique number of tokens: ', len(unique_lower_ram)/len(lower_tokens_ram))
print()

def lexical_diversity(text):
    tokens_ram = text.lower().split()
    return len(set(tokens_ram))/len(tokens_ram)
print('Lexical diversity: ', lexical_diversity(tweet_ram.text))
print()


diversities_ram = []
for tweet_ram in ramTweets:
    diversities_ram.append(lexical_diversity(tweet_ram.text))
print('Diversities: ',diversities_ram)
print('Min diversity: ', min(diversities_ram))
print()

devices_ram = []
for tweet_ram in ramTweets:
    devices_ram.append(tweet_ram.source)
print('Devices: ', devices_ram)
#print(min(diversities))


print()
print('---------------------------------------------PRIYANKA CHOPRA ---------------------------------------------------- ')
print()

tweet_priyanka = priyankaTweets[0]
print(tweet_priyanka.text)

tweet_priyanka  = priyankaTweets[0]
print(tweet_priyanka.source)

tokens_priyanka = tweet_priyanka.text.split()
print(tokens_priyanka)
print('Length: ', len(tokens_priyanka))
print()


#make everything upper or lower case
lower_tokens_priyanka = [s.lower() for s in tokens_priyanka]
print('Lower case: ', lower_tokens_priyanka)
print()


#finding out unique tokens
unique_lower_priyanka = set(lower_tokens_priyanka)
print('Unique lower tokens: ', unique_lower_priyanka)
print()
print('Unique number of tokens: ', len(unique_lower_priyanka)/len(lower_tokens_priyanka))
print()

def lexical_diversity(text):
    tokens_priyanka = text.lower().split()
    return len(set(tokens_priyanka))/len(tokens_priyanka)
print('Lexical diversity: ',lexical_diversity(tweet_priyanka.text))
print()


diversities_priyanka = []
for tweet_priyanka in priyankaTweets:
    diversities_priyanka.append(lexical_diversity(tweet_priyanka.text))
print('Diversity: ',diversities_priyanka)
print('Min diversity: ', min(diversities_priyanka))
print()


devices_priyanka = []
for tweet_priyanka in priyankaTweets:
    devices_priyanka.append(tweet_priyanka.source)
print('Devices: ',devices_priyanka)
#print(min(diversities))


print()
print('------------------------------------------------- SUNDAR PICHAI ---------------------------------------------------')
print()


tweet_sundar = sundarTweets[0]
print(tweet_sundar.text)

tweets_sundar  = sundarTweets[0]
print(tweets_sundar.source)
print()

tokens_sundar= tweets_sundar.text.split()
print(tokens_sundar)
print('Length: ', len(tokens_sundar))
print()


#make everything upper or lower case
lower_tokens_sundar = [s.lower() for s in tokens_sundar]
print('Lower case: ', lower_tokens_sundar)
print()


#finding out unique tokens
unique_lower_sundar = set(lower_tokens_sundar)
print('Unique lower tokens: ',unique_lower_sundar)
print('Unique number of tokens: ', len(unique_lower_sundar)/len(lower_tokens_sundar))
print()



def lexical_diversity(text):
    tokens_sundar = text.lower().split()
    return len(set(tokens_sundar))/len(tokens_sundar)
print('Lexical diversity: ',lexical_diversity(tweet_sundar.text))
print()


diversities_sundar = []
for tweet_sundar in sundarTweets:
    diversities_sundar.append(lexical_diversity(tweet_sundar.text))
print('Diversities: ', diversities_sundar)
print('Diversity is: ',min(diversities_sundar))
print()


devices_sundar = []
for tweet_sundar in sundarTweets:
    devices_sundar.append(tweet_sundar.source)
print('Devices: ',devices_sundar)
#print(min(diversities))

```
