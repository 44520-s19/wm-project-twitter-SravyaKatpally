```python
print()
print('----------------------------------------------RAM POTHINENI--------------------------------------------------------')
print()

tweet_ram = ramTweets[0]
print(tweet_ram.text)

tweet_ram  = ramTweets[0]
print(tweet_ram.source)
print()
tokens_ram = tweet_ram.text.split()
print('Tokens after spliting: ', tokens_ram)
print('Length: ', len(tokens_ram))
print()

#make everything upper or lower case
lower_tokens_ram = [s.lower() for s in tokens_ram]
print('Lower case: ', lower_tokens_ram)
print()

#finding out unique tokens
unique_lower_ram = set(lower_tokens_ram)
print('Unique lower tokens:  ', unique_lower_ram)
print('Unique number of tokens: ', len(unique_lower_ram)/len(lower_tokens_ram))
print()
print('-----------------LEXICAL DIVERSITY-------------------')

def lexical_diversity(text):
    tokens_ram = text.lower().split()
    return len(set(tokens_ram))/len(tokens_ram)
print('Lexical diversity: ', lexical_diversity(tweet_ram.text))
print()


diversities_ram = []
for tweet_ram in ramTweets:
    diversities_ram.append(lexical_diversity(tweet_ram.text))
print('Diversities: ',diversities_ram)
print('Min diversity: ', min(diversities_ram))
print()

devices_ram = []
for tweet_ram in ramTweets:
    devices_ram.append(tweet_ram.source)
print('Devices: ', devices_ram)


print()
print('---------------------PRIYANKA CHOPRA ---------------------------- ')
print()

tweet_priyanka = priyankaTweets[0]
print(tweet_priyanka.text)

tweet_priyanka  = priyankaTweets[0]
print(tweet_priyanka.source)

tokens_priyanka = tweet_priyanka.text.split()
print(tokens_priyanka)
print('Length: ', len(tokens_priyanka))
print()


#make everything upper or lower case
lower_tokens_priyanka = [s.lower() for s in tokens_priyanka]
print('Lower case: ', lower_tokens_priyanka)
print()


#finding out unique tokens
unique_lower_priyanka = set(lower_tokens_priyanka)
print('Unique lower tokens: ', unique_lower_priyanka)
print()
print('Unique number of tokens: ', len(unique_lower_priyanka)/len(lower_tokens_priyanka))
print()
print('-----------------LEXICAL DIVERSITY-------------------')

def lexical_diversity(text):
    tokens_priyanka = text.lower().split()
    return len(set(tokens_priyanka))/len(tokens_priyanka)
print('Lexical diversity: ',lexical_diversity(tweet_priyanka.text))
print()


diversities_priyanka = []
for tweet_priyanka in priyankaTweets:
    diversities_priyanka.append(lexical_diversity(tweet_priyanka.text))
print('Diversities: ',diversities_priyanka)
print('Min diversity: ', min(diversities_priyanka))
print()


devices_priyanka = []
for tweet_priyanka in priyankaTweets:
    devices_priyanka.append(tweet_priyanka.source)
print('Devices: ',devices_priyanka)
#print(min(diversities))


print()
print('----------------------- SUNDAR PICHAI ---------------------------')
print()


tweet_sundar = sundarTweets[0]
print(tweet_sundar.text)

tweets_sundar  = sundarTweets[0]
print(tweets_sundar.source)
print()

tokens_sundar= tweets_sundar.text.split()
print(tokens_sundar)
print('Length: ', len(tokens_sundar))
print()


#make everything upper or lower case
lower_tokens_sundar = [s.lower() for s in tokens_sundar]
print('Lower case: ', lower_tokens_sundar)
print()


#finding out unique tokens
unique_lower_sundar = set(lower_tokens_sundar)
print('Unique lower tokens: ',unique_lower_sundar)
print('Unique number of tokens: ', len(unique_lower_sundar)/len(lower_tokens_sundar))
print()


print('-----------------LEXICAL DIVERSITY-------------------')
def lexical_diversity(text):
    tokens_sundar = text.lower().split()
    return len(set(tokens_sundar))/len(tokens_sundar)
print('Lexical diversity: ',lexical_diversity(tweet_sundar.text))
print()


diversities_sundar = []
for tweet_sundar in sundarTweets:
    diversities_sundar.append(lexical_diversity(tweet_sundar.text))
print('Diversities: ', diversities_sundar)
print('Diversity is: ',min(diversities_sundar))
print()


devices_sundar = []
for tweet_sundar in sundarTweets:
    devices_sundar.append(tweet_sundar.source)
print('Devices: ',devices_sundar)
#print(min(diversities))


print()
print()
print('------------------------------------------- SENTIMENT ANALYSIS ------------------------------------------------')
import pickle
from nltk_helpers import get_sentiments

print()
print('---------------- RAM POTHINENI --------------------------')

with open('ram_pothineni.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
    
print()
print('--------------- PRIYANKA CHOPRA --------------------------')

with open('priyanka_chopra.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
    
    
print()
print('----------------- SUNDAR PICHAI --------------------------')

with open('sundar_pichai.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
```

### Reading, Analyzing and Plotting of Data : 
I have read and analyzed the data of 'Ram pothineni', 'Priyanka Chopra', 'Sundar Pichai' twitter handlers. I did both positive and negative semtiment analysis. Ram pothineni  and Priyanka Chopra both are indian actors but priyanka chopra acted in hollywood movies and she is a singer too.They tweet positively about their work. Sundar pichai is Google CEO, he tweets how google is going to be helpful to people.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('ram_pothineni.pkl', 'rb') as f:
    ram_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

ramX, ramY, ramZ = split_sentiments([get_sentiments(tweet.text) for tweet in  ram_tweets])
ax.scatter(ramX, ramY, ramZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Ram Pothineni')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('priyanka_chopra.pkl', 'rb') as f:
    priyanka_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

priyankaX, priyankaY, priyankaZ = split_sentiments([get_sentiments(tweet.text) for tweet in priyanka_tweets])
ax.scatter(priyankaX, priyankaY, priyankaZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Priyanka Chopra')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in priyanka_tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('sundar_pichai.pkl', 'rb') as f:
    sundar_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

sundarX, sundarY, sundarZ = split_sentiments([get_sentiments(tweet.text) for tweet in sundar_tweets])
ax.scatter(sundarX, sundarY, sundarZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Sundar Pichai')
plt.show()



plt.scatter(xs,zs,label='SundarPichai')
plt.scatter(priyankaX,priyankaZ,color='r',marker='*',label='PriyankaChopra')
plt.scatter(ramX,ramZ,color='g',marker='P',label='ram')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('Positive vs Negative Sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```
