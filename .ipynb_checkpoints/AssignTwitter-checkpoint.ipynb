```python
print('RAM')

tweet_ram = tweets_ram1[0]
print(tweet_ram.text)

tweet_ram  = tweets_ram1[0]
print(tweet_ram.source)

tokens_ram= tweet_ram.text.split()
print(tokens_ram)
print(len(tokens_ram))


#make everything upper or lower case
lower_tokens_ram = [s.lower() for s in tokens_ram]
print(lower_tokens_ram)


#finding out unique tokens
unique_lower_ram = set(lower_tokens_ram)
print(unique_lower_ram)
print('lexical diversity', len(unique_lower_ram)/len(lower_tokens_ram))

def lexical_diversity(text):
    tokens_ram = text.lower().split()
    return len(set(tokens_ram))/len(tokens_ram)
print(lexical_diversity(tweet_ram.text))


diversities_ram = []
for tweet_ram in tweets_ram1:
    diversities_ram.append(lexical_diversity(tweet_ram.text))
print(diversities_ram)
print(min(diversities_ram))


devices_ram = []
for tweet_ram in tweets_ram1:
    devices_ram.append(tweet_ram.source)
print(devices_ram)
#print(min(diversities))



print('Priyanaka Chopra ')

tweet_priyanka = tweets_priyanka1[0]
print(tweet_priyanka.text)

tweet_priyanka  = tweets_priyanka1[0]
print(tweet_priyanka.source)

tokens_priyanka = tweet_priyanka.text.split()
print(tokens_priyanka)
print(len(tokens_priyanka))


#make everything upper or lower case
lower_tokens_priyanka = [s.lower() for s in tokens_priyanka]
print(lower_tokens_priyanka)


#finding out unique tokens
unique_lower_priyanka = set(lower_tokens_priyanka)
print(unique_lower_priyanka)
print('lexical diversity', len(unique_lower_priyanka)/len(lower_tokens_priyanka))

def lexical_diversity(text):
    tokens_priyanka = text.lower().split()
    return len(set(tokens_priyanka))/len(tokens_priyanka)
print(lexical_diversity(tweet_priyanka.text))


diversities_priyanka = []
for tweet_priyanka in tweets_priyanka1:
    diversities_priyanka.append(lexical_diversity(tweet_priyanka.text))
print(diversities_priyanka)
print(min(diversities_priyanka))


devices_priyanka = []
for tweet_priyanka in tweets_priyanka1:
    devices_priyanka.append(tweet_priyanka.source)
print(devices_priyanka)
#print(min(diversities))





print('SUNDAR PICHAI')


tweet_sundar = tweets_sundar1[0]
print(tweet_sundar.text)

tweets_sundar  = tweets_sundar1[0]
print(tweets_sundar.source)

tokens_sundar= tweets_sundar.text.split()
print(tokens_sundar)
print(len(tokens_sundar))


#make everything upper or lower case
lower_tokens_sundar = [s.lower() for s in tokens_sundar]
print(lower_tokens_sundar)


#finding out unique tokens
unique_lower_sundar = set(lower_tokens_sundar)
print(unique_lower_sundar)
print('lexical diversity', len(unique_lower_sundar)/len(lower_tokens_sundar))




def lexical_diversity(text):
    tokens_sundar = text.lower().split()
    return len(set(tokens_sundar))/len(tokens_sundar)
print(lexical_diversity(tweet_sundar.text))


diversities_sundar = []
for tweet_sundar in tweets_ram1:
    diversities_sundar.append(lexical_diversity(tweet_sundar.text))
print(diversities_sundar)
print(min(diversities_sundar))


devices_sundar = []
for tweet_sundar in tweets_priyanka1:
    devices_sundar.append(tweet_sundar.source)
print(devices_sundar)
#print(min(diversities))

```

```python
import pickle
from nltk_helpers import get_sentiments

with open('ram_pothineni.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))

import pickle
from nltk_helpers import get_sentiments

with open('priyanka_chopra.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
    
import pickle
from nltk_helpers import get_sentiments

with open('sundar_pichai.pkl','rb') as f:
    tweets=pickle.load(f)

#print(tweets)
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[0].text))
print(get_sentiments(tweets[1].text))
for tweet in tweets:
    print(get_sentiments(tweet.text))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('ram_pothineni.pkl', 'rb') as f:
    ram_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

ramX, ramY, ramZ = split_sentiments([get_sentiments(tweet.text) for tweet in  ram_tweets])
ax.scatter(ramX, ramY, ramZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Ram Pothineni')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('priyanka_chopra.pkl', 'rb') as f:
    priyanka_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

priyankaX, priyankaY, priyankaZ = split_sentiments([get_sentiments(tweet.text) for tweet in priyanka_tweets])
ax.scatter(priyankaX, priyankaY, priyankaZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Priyanka Chopra')
plt.show()



from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

#build our figure
fig = plt.figure()
# build the axes
ax = fig.add_subplot(111, projection='3d')

sentiments = [get_sentiments(tweet.text) for tweet in tweets]

# build our xs, ys, and zs
xs = []
for sent in sentiments:
    xs.append(sent['neg'])

ys = [sent['neu'] for sent in sentiments]
zs = list(map(lambda x: x['pos'], sentiments))

ax.scatter(xs, ys, zs)

#add another set of data
import pickle
with open('sundar_pichai.pkl', 'rb') as f:
    sundar_tweets = pickle.load(f)

from nltk_helpers import split_sentiments

sundarX, sundarY, sundarZ = split_sentiments([get_sentiments(tweet.text) for tweet in sundar_tweets])
ax.scatter(sundarX, sundarY, sundarZ, color='r', marker='^')

ax.set_xlabel('Negative')
ax.set_ylabel('Neutral')
ax.set_ylabel('Positive')
plt.title('Sundar Pichai')
plt.show()



plt.scatter(xs,zs,label='Ram')
plt.scatter(priyankaX,priyankaZ,color='r',marker='$\clubsuit$',label='PriyankaChopra')
plt.scatter(sundarX,sundarZ,color='g',marker='d',label='SundarPichai')
plt.xlabel('negative')
plt.ylabel('positive')
plt.title('Positive vs Negative Sentiments')
plt.legend(loc='upper right',shadow=True)
plt.show()
```
